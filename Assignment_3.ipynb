```python
#IMPORTS
import gymnasium as gym
import minatar
import minatar.gym
import random
from minatar import Environment
```

```python
env = Environment('breakout')
num_actions = env.num_actions()
print(env.game_name(), env.n_channels)

# Instead of env.display_state, use matplotlib to display one channel as an image
import matplotlib.pyplot as plt
import numpy as np

def show_env_state(state):
    # State shape: (channels, height, width); show sum over channels or pick a channel
    if isinstance(state, np.ndarray):
        img = np.sum(state, axis=0)  # collapse channels for visualization
    else:
        img = np.sum(np.array(state), axis=0)
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.show()

G = 0
terminated = False
state = env.state()
show_env_state(state)
while not terminated:
    action = random.randrange(num_actions)
    print("taking action", action)
    reward, terminated = env.act(action)
    s_prime = env.state()
    G += reward
    show_env_state(s_prime)

print(G)
```
